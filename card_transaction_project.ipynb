```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
data_1 = pd.read_csv('card_transactions.csv')
```

```python
data_2 = data_1.drop(['Unnamed: 0', 'payment_id', 'client_id', 'transaction_date', 'country_code', 'network_dist'], axis=1)
```

```python
fraud_equal_1=data_2.loc[data_2['fraud']==1]
```

```python
fraud_equal_1
```

```python
new_data = pd.concat([data_2.iloc[:12000], fraud_equal_1])
new_data
```

```python
new_1_data = pd.concat([new_data, pd.get_dummies(new_data.mcc), pd.get_dummies(new_data.card_type), pd.get_dummies(new_data.proxy_ip)], axis=1)
new_1_data
```

```python
new_2_data = new_1_data.drop(['card_scheme', 'mcc', 'card_type', 'proxy_ip', 'battery_score', 'bh_score'], axis =1)
```

```python
from sklearn.model_selection import train_test_split
y = new_2_data['fraud']
x = new_2_data.drop(['fraud'], axis=1)
x_train, x_valid, y_train, y_valid = train_test_split(x, y, train_size=0.8)
```

**Shu bolekde simple imputer ulanyp ishledim**

```python
from sklearn.impute import SimpleImputer

sim_imputer = SimpleImputer()
xs_train = pd.DataFrame(sim_imputer.fit_transform(x_train))
xs_valid = pd.DataFrame(sim_imputer.transform(x_valid))

xs_train.columns = x_train.columns
xs_valid.columns = x_valid.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
model_1 = RandomForestRegressor(random_state=0)
model_1.fit(xs_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre= model_1.predict(xs_valid)
pred = []
for i in pre:
    if i<0.5 :
        pred.append(0)
    elif i>=0.5:
        pred.append(1)
res = np.array(pred)
accuracy_score(res, y_valid)
```

**Bu bolekde missing value doldurmagyng 3nji usulyny ulandym**

```python
cols_with_missing = [col for col in x_train.columns
                     if x_train[col].isnull().any()]
```

```python
xb_train = x_train.copy()
xb_valid = x_valid.copy()

for col in cols_with_missing:
    xb_train[col + '_was_missing'] = xb_train[col].isnull()
    xb_valid[col + '_was_missing'] = xb_valid[col].isnull()

bet_imputer = SimpleImputer()
xbi_train = pd.DataFrame(bet_imputer.fit_transform(xb_train))
xbi_valid = pd.DataFrame(bet_imputer.transform(xb_valid))

xbi_train.columns = xb_train.columns
xbi_valid.columns = xb_valid.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
model_2 = RandomForestRegressor(random_state=0)
model_2.fit(xbi_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre= model_2.predict(xbi_valid)
pred = []
for i in pre:
    if i<0.5 :
        pred.append(0)
    elif i>=0.5:
        pred.append(1)
res = np.array(pred)
accuracy_score(res, y_valid)
```
